```python
def test_cases_init():
    ranges = [10,20,50,100,150,200,250,300]
    presorted = []
    reversedsorted = []

    for r in ranges:
        i = 1
        arr = []
        while (i <= r):
            arr.append(i)
            i +=1
        presorted.append(arr)
        reversedsorted.append(arr[::-1])

    return presorted, reversedsorted
```
